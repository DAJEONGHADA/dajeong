```python
import requests

# 카카오 API 키
kakao_api_key = "48dc83d22ece1e3192112cee2bbd22eb"

def address(place_address):

    url = 'https://dapi.kakao.com/v2/local/search/address.json'
    headers = {'Authorization': f'KakaoAK {kakao_api_key}'}
    params = {'query': place_address}

    response = requests.get(url, headers=headers, params=params)
    
    if response.status_code == 200:
        json_result = response.json()
        if json_result["documents"]:
            address_info = json_result["documents"][0]
            return address_info["x"], address_info["y"]
        else:
            print("검색 결과가 없습니다.")
    else:
        print(f"HTTP 오류 발생: {response.status_code} - {response.reason}")
    return None, None
```

```python
from math import sin, cos, sqrt, atan2, radians

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # radius of Earth in kilometers

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance
```

```python
my_address = input("현재 도로명 주소를 입력하세요: ")
lon, lat = address(my_address)

print(f"경도: {lon}, 위도: {lat}")
# 127. ... , 37...
lon1 = float(lon)
lat1 = float(lat)
```

```python
distance_answer = []

for lat2, lon2 in zip(food_filter["위도"], food_filter["경도"]):
    dist = haversine_distance(lat1, lon1, lat2, lon2)
    distance_answer.append(dist)

food_filter["distance_km"] = distance_answer
# food_filter
# food_filter.info()
```

```python
food_km = food_filter[food_filter["distance_km"] < 10]
food_km
```

#### 6. 선택되었으면 folium 을 이용하여 map을 만든다.

```python
import folium
from folium import Marker, Icon, CircleMarker
import random


center = [lat, lon]
food_map = folium.Map(location=center, zoom_start=12)

folium.Marker(location=center, popup="내 위치", icon=folium.Icon(color="red")).add_to(food_map)
folium.Circle(
    location=center,
    radius=10000,  # meter 단위 (10km = 10000m)
    color="tomato",
    fill=True,
    fill_color="tomato",
    fill_opacity=0.3,
    tooltip="내 위치에서 10km 이내 표시"
).add_to(food_map)

for name, lat_map, lon_map in zip(food_km["식당상호"], food_km["위도"], food_km["경도"]):
    folium.Marker([lat_map, lon_map], popup=name).add_to(food_map)
food_map

지도를 HTML 파일로 저장하기
food_map.save("D:/멀티캠퍼스/010일차_(240612)/food_map.html")

# food_map.save("http://www.food_map.html")


```
